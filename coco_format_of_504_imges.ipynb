```python
import os
import json
import random
from PIL import Image, ImageDraw, ImageEnhance
```

```python


# output_dir = 'images/7th_rotated_img/'
file = {"info": {}, "licenses": [], "images": [], "annotations": [], "categories": [{"id": 1, "name": "dog"}, {"id": 2, "name": "cat"}]}
with open("cat_dog_annotations.json") as f:
    json_data = json.load(f)
    data = json_data["_via_img_metadata"].values()
    for idx, img in enumerate(data):
        image_data = {"id": idx + 1, "file_name": img["filename"], "height": 0, "width": 0}
        file["images"].append(image_data)
        img_file = Image.open(img["filename"])
        for region in img["regions"]:
            x = region['shape_attributes']['x']
            y = region['shape_attributes']['y']
            w = region['shape_attributes']['width']
            h = region['shape_attributes']['height']
            label = list(region["region_attributes"]["class"].keys())[0]
            new_img = img_file.resize((170, 100))
            draw = ImageDraw.Draw(new_img)
            draw.rectangle((x, y, x + w, y + h), outline='blue')
            draw.text((x, y - 8), label, fill='red')
            box_size = 85
            center_x = new_img.size[0] / 2
            center_y = new_img.size[1] / 2
            box_coords = (center_x - box_size / 2, center_y - box_size / 2,
                          center_x + 15 + box_size / 2, center_y + box_size / 2)
            draw.rectangle(box_coords, outline="red", width=2)
            brightness = random.uniform(1, 2)
            enhancer = ImageEnhance.Brightness(new_img)
            img_mod = enhancer.enhance(brightness)
            for angle in range(0, 360, 5):
                rotate_img = img_mod.rotate(angle)
                file_name = f"rotate_img_{idx + 1}_{angle}.jpeg"
#                 rotate_img.save(os.path.join(output_dir, file_name))
                # plt.imshow(rotate_img)
                # plt.show()
                annotation_data = {"id": len(file["annotations"]) + 1, "image_id": idx + 1,
                                   "category_id": 0, "bbox": [x, y, w, h]}
                if "dog" in region["region_attributes"]["class"]:
                    annotation_data["category_id"] = 1
                elif "cat" in region["region_attributes"]["class"]:
                    annotation_data["category_id"] = 2
                file["annotations"].append(annotation_data)

with open("output.json1", "w") as f:
    json.dump(file, f)
```
